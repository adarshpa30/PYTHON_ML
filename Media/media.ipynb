```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"D:\Global YouTube Statistics.csv", encoding='latin1')
df
```

```python
print(df.info())
print(df.describe())
```

```python
df = df.dropna() 
df['subscribers'] = df['subscribers'].astype(int)
df['video views'] = df['video views'].astype(int)
df['uploads'] = df['uploads'].astype(int)
df['created_year'] = pd.to_datetime(df['created_year'], format='%Y')
print(df.info())
```

```python
#Top 10 YouTube channels based on the number of subscribers:
top_10_channels = df.nlargest(10, 'subscribers')[['Youtuber', 'subscribers']]
print(top_10_channels)
```

```python
#Category with the highest average number of subscribers:
category_avg_subs = df.groupby('category')['subscribers'].mean().idxmax()
print(category_avg_subs)
```

```python
#Average number of videos uploaded by YouTube channels in each category:
category_avg_uploads = df.groupby('category')['uploads'].mean()
print(category_avg_uploads)
```

```python
#Top 5 countries with the highest number of YouTube channels:
top_5_countries = df['Country'].value_counts().head(5)
print(top_5_countries)
```

```python
#Distribution of channel types across different categories
channel_type_dist = df.groupby(['category', 'Title']).size()
print(channel_type_dist)
```

```python
#Correlation between the number of subscribers and total video views:
correlation_subs_views = df['subscribers'].corr(df['video views'])
print(correlation_subs_views)
```

```python
#Trend in subscribers gained in the last 30 days across all channels:
trend_subs_last_30_days = df['subscribers_for_last_30_days'].sum()
print(trend_subs_last_30_days)
```

```python
#Relationship Between Gross Tertiary Education Enrollment and Number of YouTube Channels
channel_counts = df['Country'].value_counts().reset_index()
channel_counts.columns = ['Country', 'Number_of_Channels']
education_enrollment = df.groupby('Country')['Gross tertiary education enrollment (%)'].mean().reset_index()
merged_data = pd.merge(channel_counts, education_enrollment, on='Country')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_data, x='Gross tertiary education enrollment (%)', y='Number_of_Channels')
plt.xlabel('Gross Tertiary Education Enrollment (%)')
plt.ylabel('Number of YouTube Channels')
plt.title('Relationship Between Gross Tertiary Education Enrollment and Number of YouTube Channels')
plt.show()
```

```python
#Unemployment Rate Variation Among Top 10 Countries with the Highest Number of YouTube Channels
top_10_countries = df['Country'].value_counts().head(10).index
top_10_df = df[df['Country'].isin(top_10_countries)]
plt.figure(figsize=(12, 6))
sns.boxplot(data=top_10_df, x='Country', y='Unemployment rate')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate in Top 10 Countries with Highest Number of YouTube Channels')
plt.show()
```

```python
# Average Urban Population Percentage in Countries with YouTube Channels
average_urban_population = df['Urban_population'].mean()
print(f"Average urban population percentage: {average_urban_population:.2f}%")
```

```python
# Patterns in Distribution of YouTube Channels Based on Latitude and Longitude Coordinates
plt.figure(figsize=(12, 8))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribution of YouTube Channels Based on Latitude and Longitude')
plt.show()
```

```python
#Correlation Between Number of Subscribers and Population of a Country
# Calculate correlation
correlation_subs_population = df['subscribers'].corr(df['Population'])
print(f"Correlation between number of subscribers and population: {correlation_subs_population:.2f}")
```

```python
#Comparison of Total Population in Top 10 Countries with the Highest Number of YouTube Channels
population_data = df.groupby('Country')['Population'].mean().reset_index()
top_10_population = population_data[population_data['Country'].isin(top_10_countries)]
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_population, x='Country', y='Population')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Total Population of Top 10 Countries with Highest Number of YouTube Channels')
plt.show()
```

```python
#Correlation Between Subscribers Gained in Last 30 Days and Unemployment Rate
correlation_subs_30_days_unemployment = df['subscribers_for_last_30_days'].corr(df['Unemployment rate'])
print(f"Correlation between subscribers gained in last 30 days and unemployment rate: {correlation_subs_30_days_unemployment:.2f}")
```
