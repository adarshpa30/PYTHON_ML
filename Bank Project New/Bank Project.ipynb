```python
df=pd.get_dummies(df).astype(int)
df
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score,roc_auc_score
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
x=df.drop(['y'],axis=1)
y=df['y']
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=24)
```

```python
xgb_classifier = xgb.XGBClassifier()

xgb_classifier.fit(X_train, y_train)

y_pred = xgb_classifier.predict(X_test)
y_pred_proba = xgb_classifier.predict_proba(X_test)[:, 1]

auc = roc_auc_score(y_test, y_pred_proba)
precision = precision_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print(f"XGB AUC: {auc:.4f}")
print(f"XGB Precision: {precision:.4f}")
print(f"XGB Accuracy: {accuracy:.4f}")
```

```python
rf_classifier = RandomForestClassifier(random_state=42)

rf_classifier.fit(X_train, y_train)

y_pred_rf = rf_classifier.predict(X_test)
y_pred_proba_rf = rf_classifier.predict_proba(X_test)[:, 1]

auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
precision_rf = precision_score(y_test, y_pred_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(f"Random Forest - AUC: {auc_rf:.4f}")
print(f"Random Forest - Precision: {precision_rf:.4f}")
print(f"Random Forest - Accuracy: {accuracy_rf:.4f}")
```

```python
nb_classifier = GaussianNB()

nb_classifier.fit(X_train, y_train)

y_pred_nb = nb_classifier.predict(X_test)
y_pred_proba_nb = nb_classifier.predict_proba(X_test)[:, 1]

auc_nb = roc_auc_score(y_test, y_pred_proba_nb)
precision_nb = precision_score(y_test, y_pred_nb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print(f"Naive Bayes - AUC: {auc_nb:.4f}")
print(f"Naive Bayes - Precision: {precision_nb:.4f}")
print(f"Naive Bayes - Accuracy: {accuracy_nb:.4f}")
```

```python
dt_classifier = DecisionTreeClassifier(random_state=42)

dt_classifier.fit(X_train, y_train)

y_pred_dt = dt_classifier.predict(X_test)
y_pred_proba_dt = dt_classifier.predict_proba(X_test)[:, 1]

auc_dt = roc_auc_score(y_test, y_pred_proba_dt)
precision_dt = precision_score(y_test, y_pred_dt)
accuracy_dt = accuracy_score(y_test, y_pred_dt)

print(f"Decision Tree - AUC: {auc_dt:.4f}")
print(f"Decision Tree - Precision: {precision_dt:.4f}")
print(f"Decision Tree - Accuracy: {accuracy_dt:.4f}")
```

```python
results = {
    "Random Forest": {"AUC": auc_rf, "Precision": precision_rf, "Accuracy": accuracy_rf},
    "Naive Bayes": {"AUC": auc_nb, "Precision": precision_nb, "Accuracy": accuracy_nb},
    "Decision Tree": {"AUC": auc_dt, "Precision": precision_dt, "Accuracy": accuracy_dt},
    "XGB":{"AUC":auc,"Precision":precision,"Accuracy":accuracy}
}
```

```python
models = list(results.keys())
metrics = list(results[models[0]].keys())
n_metrics = len(metrics)


fig, axes = plt.subplots(n_metrics, 1, figsize=(10, 8), sharex=True)

colors = ["#FF5733", "#FFA500", "#FFC300", "#8D33FF", "#33FFF2"]

for i, metric in enumerate(metrics):
    values = [results[model][metric] for model in models]
    axes[i].bar(models, values, color=colors[:len(models)])
    axes[i].set_title(f'{metric} for different models')
    axes[i].set_ylabel(metric)
    axes[i].set_ylim(0, 1)

axes[-1].set_xlabel('Models')
plt.tight_layout()
plt.show()
```

"The XGBoost model is the best choice due to its highest AUC, indicating superior overall performance in distinguishing between clients who will subscribe to a term deposit and those who won't. Despite having slightly lower precision than the Random Forest model, its overall effectiveness makes it the most suitable model for this classification task"
