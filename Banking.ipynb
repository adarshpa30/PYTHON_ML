```python
#import libaries:
import pandas as pd
import numpy as  np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#load data set:
data=pd.read_csv("D:/SEM 4/bank-additional-full.csv")
```

```python
data
```

```python
data.head()
```

```python
#b. Print the information about the dataset:
data.info()
```

```python
#c. Pre-process the dataset. Remove missing values.:
data.isnull().sum()
```

```python
data['marital'].fillna(data['marital'].mode()[0],inplace=True)
data['education'].fillna(data['education'].mode()[0],inplace=True)
data['default'].fillna(data['default'].mode()[0],inplace=True)
data['education'].fillna(data['education'].mode()[0],inplace=True)
data['housing'].fillna(data['housing'].mode()[0],inplace=True)
data['loan'].fillna(data['loan'].mode()[0],inplace=True)
data['contact'].fillna(data['contact'].mode()[0],inplace=True)
data['month'].fillna(data['month'].mode()[0],inplace=True)
data['day_of_week'].fillna(data['day_of_week'].mode()[0],inplace=True)
data['poutcome'].fillna(data['poutcome'].mode()[0],inplace=True)
data['duration'].fillna(data['duration'].mean(),inplace=True)
data['campaign'].fillna(data['campaign'].mean(),inplace=True)
data['pdays'].fillna(data['pdays'].mean(),inplace=True)
data['previous'].fillna(data['previous'].mean(),inplace=True)
data['emp.var.rate'].fillna(data['emp.var.rate'].mean(),inplace=True)
data['cons.price.idx'].fillna(data['cons.price.idx'].mean(),inplace=True)
data['cons.conf.idx'].fillna(data['cons.conf.idx'].mean(),inplace=True)
data['euribor3m'].fillna(data['euribor3m'].mean(),inplace=True)
data['nr.employed'].fillna(data['nr.employed'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
print('values:',data['job'].unique())
print('values:',data['marital'].unique())
print('values:',data['education'].unique())
print('values:',data['default'].unique())
print('values:',data['housing'].unique())
print('values:',data['loan'].unique())
print('values:',data['contact'].unique())
print('values:',data['month'].unique())
print('values:',data['day_of_week'].unique())
print('values:',data['poutcome'].unique())
```

```python
#converting:
data['job']=data['job'].map({'housemaid':0,'services':1, 'admin.':2,'blue-collar':3,'technician':4,'retired':5,'management':6,'unemployed':7,'self-employed':8,'unknown':9,'entrepreneur':10,'student':11})
data['marital']=data['marital'].map({'married':1,'single':0,'divorced':2,'unknown':3})
data['education']=data['education'].map({'basic.4y':0,'high.school':1,'basic.6y':2 ,'basic.9y':3 ,'professional.course':4,'unknown':5,'university.degree':6,'illiterate':7})
data['default']=data['default'].map({'no':0,'unknown':2,'yes':1})
data['housing']=data['housing'].map({'no':0,'yes':1,'unknown':2})
data['loan']=data['loan'].map({'no':0,'yes':1,'unknown':2})
data['contact']=data['contact'].map({'telephone':0,'cellular':1})
data['month']=data['month'].map({'may':5,'jun':6,'jul':7,'aug':8,'oct':10,'nov':11,'dec':12,'mar':3,'apr':4,'sep':8})
data['day_of_week']=data['day_of_week'].map({'mon':0,'thu':1,'tue':3,'wed':2,'fri':4})
data['poutcome']=data['poutcome'].map({'nonexistent':0,'failure':1,'success':2})
```

```python
data.head()
```

```python
# Feature Scaling:
n=data.iloc[:,:-1]
data.iloc[:,:-1]=(n-n.min())/(n.max()-n.min())
```

```python
data
```

```python
#a. Print feature values and target values:
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
#d.Reduce the features to 5 components using pca:
pca=PCA()
pca=PCA(n_components=5)
x=pca.fit_transform(x)
```

```python
x
```

```python
#d.Explain lazy learning algorithm:
#KNN is Called lazy learner knn didn't learn from the discrtive function
```

```python
#training and testing:
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
#e. Using the lazy learning algorithm create a model which predicts the future results:
classifier=KNeighborsClassifier(n_neighbors=203,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
predict=classifier.predict(x_test)
```

```python
predict
```

```python
#f. Find the accuracy of the model.
print(metrics.accuracy_score(predict,y_test))
```

```python
#g. Print the classification report:
print(classification_report(predict,y_test))
```

```python
##h.Explain confusion matrix using above model:
cm=confusion_matrix(y_test,predict)
print('matrix:',cm)
```
