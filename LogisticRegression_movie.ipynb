```python
#importng the libraies:
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
#loading dataset:
movie=pd.read_csv("D:\SEM 4\Movie_classification.csv")
```

```python
movie
```

```python
#dataset information:
movie.info()
```

```python
print(movie['Genre'].unique())
print(movie['3D_available'].unique())
```

```python
#cathegorical to numerical:
movie['Genre']=movie['Genre'].map({'Thriller':0,'Drama':1,'Comedy':2,'Action':3})
movie['3D_available']=movie['3D_available'].map({'YES':1,'NO':0})
```

```python
movie.head()
```

```python
#checkig the null values:
movie.isnull().sum()
```

```python
#changing the null values with mean:
movie['Time_taken'].fillna(movie['Time_taken'].mean(),inplace=True)
```

```python
movie.isnull().sum()
```

```python
# Feature Scaling:
n=movie.iloc[:,:-1]
movie.iloc[:,:-1]=(n-n.min())/(n.max()-n.min())
```

```python
movie
```

```python
#spliting dataset into feature and traget values:
x=movie.iloc[:,:-1].values
y=movie.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
#reducing the dataset 3 components:
pca=PCA()
pca=PCA(n_components=3)
```

```python
x=pca.fit_transform(x)
```

```python
x
```

```python
#Splitting the dataset into the Training set and Test set:
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(x_test))
```

```python
y_train
```

```python
# Training the Logistic Regression model:
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
# Predicting the Test set results:
predict=lg.predict(x_test)
```

```python
predict
```

```python
#prtinting the accuracy score:
print('accuracy score:',metrics.accuracy_score(y_test,predict))
```

```python
#priting the classification report:
print(classification_report(y_test,predict))
```
