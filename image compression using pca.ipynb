```python
#loading libraeis
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#loading image:("D:\personal\photos\IMG_20200728_210428.jpg")
img=cv2.imread
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
#splitting the image into RGB
blue,green,red=cv2.split(img)
```

```python
blue.shape
```

```python
plt.imshow(blue)
```

```python
plt.imshow(green)
```

```python
plt.imshow(red)
```

```python
pca=PCA(n_components=20)
```

```python
red_transform=pca.fit_transform(red)
green_transform=pca.fit_transform(green)
blue_transform=pca.fit_transform(blue)
```

```python
red.shape
```

```python
red_transform.shape
```

```python
green_transform.shape
```

```python
blue_transform.shape
```

```python
green_transform
```

```python
#inversing the transforming:
red_inverse=pca.inverse_transform(red_transform)
green_inverse=pca.inverse_transform(green_transform)
blue_inverse=pca.inverse_transform(blue_transform)
```

```python
red_inverse
```

```python
green_inverse
```

```python
blue_inverse
```

```python
img_compressed=(np.dstack((blue_inverse,green_inverse,red_inverse))).astype(np.uint8)
```

```python
plt.imshow(img_compressed)
```

```python
cv2.imwrite('D:\joey1.jpeg',img_compressed)
```
